```python
import scanpy as sc
adata = sc.read_h5ad("data/sound-life_whole-blood_no-stim.h5ad")
adata.X = adata.X.astype('float32')

sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

sc.pp.highly_variable_genes(adata, n_top_genes=1000, flavor='seurat')

sc.pl.highly_variable_genes(adata)
```

I implemented an option to select the n most variables genes from the dataset so that the notebook can be run either with that option on None (all genes), or on a selected number. It seems that the most selective we are on genes, the worse the MSE. I suppose lots of genes are mostly constant and therefore easy to predict, thus better results with all genes. 

Nevertheless, it is not what I expected at first, as I thought selecting only highly variable genes would yield better results as they would be the most explanable, and it would remove noise
